```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
employee_info = pd.read_csv('D:/old/Employees_info/2009-12.csv')
employee_info.columns = ['employee_name', 'user_id', 'Domain', 'Email', 'Role']
Role_data=employee_info.groupby(['Role'])['user_id'].count()
Role_data.columns = ['Role','Count']
fig,ax = plt.subplots(1,1,figsize=(20,5))
#Role_data.plot(ax=ax)
labels=list(Role_data.index)
plt.bar(range(len(Role_data)), Role_data)
plt.xticks(range(len(Role_data)),labels,size='large',rotation=60)
plt.ylabel('Number of People')
plt.xlabel('Role')
plt.title('Distribution of Role')
plt.show()
```

```python
psycho_data = pd.read_csv("D:/new/psychometric_info.csv")
user_id_30_50=employee_info[psycho_data["O"]>30]
Role_data=user_id_30_50.groupby(['Role'])['user_id'].count()
Role_data.columns = ['Role','Count']
fig,ax = plt.subplots(1,1,figsize=(20,5))
labels=list(Role_data.index)
plt.bar(range(len(Role_data)), Role_data)
plt.xticks(range(len(Role_data)),labels,size='large',rotation=60)
plt.ylabel('Number of People')
plt.xlabel('Role')
plt.title('Distribution of Role')
plt.show()
```
