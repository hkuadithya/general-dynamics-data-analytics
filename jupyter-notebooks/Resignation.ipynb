# which role/functional_unit/department/team/supervisor has the maximum amount of resignation

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
emp=pd.read_pickle('emp_info_last_access')
emp
```

# acquire information about who has left

```python
resignation=emp[emp['left']==1]
resignation
```

```python
Role_resignation=resignation.groupby(['role'])['user_id'].count()
Role_resignation.columns = ['Role','Count']
fig,ax = plt.subplots(1,1,figsize=(20,5))
labels=list(Role_resignation.index)
plt.bar(range(len(Role_resignation)), Role_resignation)
plt.xticks(range(len(Role_resignation)),labels,size='large',rotation=60)
plt.ylabel('Number of Resignation')
plt.xlabel('Role')
#plt.title('')
plt.show()
```

```python
FU_resignation=resignation.groupby(['functional_unit'])['user_id'].count()
FU_resignation.columns = ['functional_unit','Count']
fig,ax = plt.subplots(1,1,figsize=(10,5))
labels=list(FU_resignation.index)
plt.bar(range(len(FU_resignation)), FU_resignation)
plt.xticks(range(len(FU_resignation)),labels,size='large',rotation=60)
plt.ylabel('Number of Resignation')
plt.xlabel('functional_unit')
#plt.title('')
plt.show()
```

```python
Dep_resignation=resignation.groupby(['department'])['user_id'].count()
Dep_resignation.columns = ['Department','Count']
fig,ax = plt.subplots(1,1,figsize=(20,5))
labels=list(Role_resignation.index)
plt.bar(range(len(Dep_resignation)), Dep_resignation)
plt.xticks(range(len(Dep_resignation)),labels,size='large',rotation=60)
plt.ylabel('Number of Resignation')
plt.xlabel('Department')
#plt.title('')
plt.show()
```

```python
Team_resignation=resignation.groupby(['team'])['user_id'].count()
Team_resignation.columns = ['Team','Count']
fig,ax = plt.subplots(1,1,figsize=(25,5))
labels=list(Team_resignation.index)
plt.bar(range(len(Team_resignation)), Team_resignation)
plt.xticks(range(len(Team_resignation)),labels,size='large',rotation=90)
plt.ylabel('Number of Resignation')
plt.xlabel('Team')
#plt.title('')
plt.show()
```
