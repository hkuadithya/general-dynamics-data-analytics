```python
emp_final['sentiment']=np.where(emp_final['content_y']>=0, 1, 0)
```

```python
from sklearn.preprocessing import LabelEncoder

emp_final1=emp_final
lb = LabelEncoder()
emp_final1['role'] = lb.fit_transform(emp_final1['role'])
emp_final1['functional_unit'] = lb.fit_transform(emp_final1['functional_unit'].fillna('0'))
emp_final1['department'] = lb.fit_transform(emp_final1['department'].fillna('0'))
emp_final1['supervisor'] = lb.fit_transform(emp_final1['supervisor'].fillna('0'))
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score

features=['role','functional_unit','department','supervisor','content_y','pos_count','neg_count','N','O','C','E','A']

X=emp_final1[features]
Y = emp_final1["isSuspicious"]
train_x, test_x, train_y, test_y  =   train_test_split(X, Y, test_size=.3)
trained_model = RandomForestClassifier()
trained_model.fit(train_x, train_y)
feature_importance=trained_model.feature_importances_
features = np.asarray(features)
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) 
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, features[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

```python
predictions=trained_model.predict(test_x)
print(confusion_matrix(test_y,predictions))
print(accuracy_score(test_y, predictions))
```
